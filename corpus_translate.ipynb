```python
from deep_translator.validate import NotValidPayload
import re 
import difflib


def has_cyrillic(text):
  return bool(re.search('[а-яА-Я]', text))

def has_numbers(inputString):
    return any(char.isdigit() for char in inputString)

def translate(original):
  return translator.translate(original, dest='ru', src='en')

def tr_roots(row):
  if has_cyrillic(row[1][0]['span_text'][0]):
    return row[1]

  else:
    try:
      for span in row[1]:
        target = translate(span['span_text'])
        sentence = nlp(row[0])
        sentence_list = [w.text for w in sentence]
        word = difflib.get_close_matches(target, sentence_list, n=1)

        if word:
          index = [i for i,x in enumerate(sentence_list) if x == word[0]][0]
          span1 = [index, index + 1]
          span['span1'], span['span_text'] = span1, word[0]

        else:
          span['span1'] = None
          
    except (NotValidPayload, TypeError):
      pass

    return row[1]


df['targets'] = df.apply(tr_roots, axis=1)
```

```python
def roots(row):
  id = 0
  roots = {}
  for token in nlp(row[0]):
    if token.dep_ == 'ROOT' or token.dep_ == 'ccomp':
      roots[token.text] = id 
    id += 1

  if len(roots) == len(row[1]):
    target_id = 0
    for root in roots:
      span = [roots[root], roots[root] + 1]
      row[1][target_id]['span1'], row[1][target_id]['span_text'] = span, root
      target_id += 1

  return row[1]

df['targets'] = df.apply(roots, axis=1)
```
